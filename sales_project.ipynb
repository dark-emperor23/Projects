```python
ax = sb.countplot(x ='Gender',data = df)
for bars in ax.containers :
    ax.bar_label(bar) #gender count
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending =False)
sb.barplot(x='Gender',y='Amount',data= sales_gen) #GENDER WISE SALES
```

```python
ax = sb.countplot(data = df, x = 'Age Group', hue = 'Gender') #age group count
```

```python
sb.barplot(x= 'Age Group',y ='Amount', data = df) #Sales by age-group
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sb.set(rc={'figure.figsize':(20,6)})
sb.barplot(data = sales_state,x='State',y='Orders')
#State-wise Orders
```

```python
  sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sb.set(rc={'figure.figsize':(20,6)})
sb.barplot(data=sales_state, x ='Occupation',y='Amount') #Amount Spent State-wise
```

```python
sb.set(rc={'figure.figsize':(25,10)})
ax= sb.countplot(data = df, x = 'Product_Category') #Comparison of occupations sending money
```

```python
sb.set(rc={'figure.figsize':(25,10)})
ax= sb.countplot(data = df, x = 'Marital_Status') #Comparison of Marital Status of people spending money
```

```python
sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(25,10)})
ax = sb.barplot(data = sales_state,x='Product_Category',y='Amount') #Comparison of products were money is spent
```

```python
sales_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.set(rc={'figure.figsize':(25,10)})
ax = sb.barplot(data = sales_state,x='Product_ID',y='Orders') #Sales of the product
```

```python
fig,ax = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
 #comparison of each products according to product id                                                                                       
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sb.set(rc={'figure.figsize':(20,6)})
sb.barplot(data = sales_state,x='State',y='Orders')
#State-wise comparison of orders received
```
